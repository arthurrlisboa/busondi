## Ônibus

```python
pasta = 'static/'
arquivo = 'routes.csv'

routes = pd.read_csv(pasta+arquivo)
routes
```

```python
pasta = 'static/'
arquivo = 'trips.csv'

trips = pd.read_csv(pasta+arquivo)
trips
```

```python
pasta = 'static/'
arquivo = 'stop_times.csv'

stop_times = pd.read_csv(pasta+arquivo)
stop_times
```

```python
# Remove as duplicadas pelo route_id
temp = trips.drop_duplicates(subset='route_id').set_index('route_id')

# Join de routes com trips
routes_trips = routes.join(temp, on='route_id', how='inner')
routes_trips
```

```python
# Seleciona apenas o primeiro e último ponto da viagem
temp = stop_times.dropna(subset=['departure_time']).set_index('trip_id')

# Join de routes e trips com stop_times
routes_trips_stop_times = routes_trips.join(temp, on='trip_id', how='inner')
routes_trips_stop_times
```

```python
# Faz um shift dos pontos, para poder adicionar o ponto final na mesma linha do inicial
temp = list(routes_trips_stop_times['stop_id'][1:].append(pd.Series([0])))
routes_trips_stop_times['final_stop_id'] = temp

# Remove as duplicatas no trip_id
routes_trips_stop_times = routes_trips_stop_times.drop_duplicates(subset='trip_id')
routes_trips_stop_times
```

```python
# Seleciona colunas relevantes
onibus = routes_trips_stop_times[['route_id', 'route_short_name', 'route_long_name', 'shape_id', 'stop_id', 'final_stop_id']]
# Renomeia as colunas
onibus = onibus.rename(columns={'stop_id' : 'initial_stop_id'})
# Descarta entradas duplicadas e vazias
onibus = onibus.drop_duplicates().dropna()
onibus
```

```python
pasta = 'processed/'
arquivo = 'routes.csv'

onibus.to_csv(pasta+arquivo, index=False)
```

## LinhaPonto

```python
temp = routes_trips_stop_times.set_index('trip_id')

# Join com a tabela completa para obter a informação de route_id das mesmas trips
stop_times_joined = stop_times.join(temp, on='trip_id', how='inner', lsuffix='_')
stop_times_joined
```

```python
# Cria um índice
stop_times_joined['route_stop_id'] = stop_times_joined.reset_index().index
# Seleciona colunas relevantes
linha_ponto = stop_times_joined[['route_stop_id', 'route_id', 'stop_id_', 'stop_sequence_']]
# Renomeia as colunas
linha_ponto = linha_ponto.rename(columns={'stop_id_' : 'stop_id',
                                          'stop_sequence_' : 'stop_sequence'})
# Descarta entradas duplicadas e vazias
linha_ponto = linha_ponto.drop_duplicates().dropna()
linha_ponto
```

```python
pasta = 'processed/'
arquivo = 'route_stop.csv'

linha_ponto.to_csv(pasta+arquivo, index=False)
```

## Partidas Ônibus

```python
# Trips válidas (selecionadas na tabela ônibus)
trips_valid_shapes = trips[trips['shape_id'].isin(onibus['shape_id'])]
trips_valid_shapes
```

```python
# Join dos stop_times com as rotas válidas para filtrar apenas os horários válidos
stop_times_valid_trips = stop_times.join(trips_valid_shapes.set_index('trip_id'), on='trip_id', how='right', lsuffix='_')
stop_times_valid_trips
```

```python
# Filtrando apenas o primeiro ponto
first_stop = stop_times_valid_trips[stop_times_valid_trips['stop_sequence'] == 1]
first_stop
```

```python
# Seleciona colunas relevantes
partidas_onibus = first_stop[['trip_id', 'departure_time', 'stop_id', 'route_id']]
# Descarta entradas duplicadas e vazias
partidas_onibus = partidas_onibus.drop_duplicates().dropna()
partidas_onibus
```

```python
pasta = 'processed/'
arquivo = 'bus_departures.csv'

partidas_onibus.to_csv(pasta+arquivo, index=False)
```

## Shapes

```python
import geopandas as gpd
```

```python
pasta = 'static/'
arquivo = 'shapes.csv'

shapes = gpd.read_file(pasta+arquivo)
shapes
```

```python
pasta = 'processed/'
arquivo = 'routes.csv'

onibus = pd.read_csv(pasta+arquivo)
onibus
```

```python
# Shapes válidas (selecionadas na tabela ônibus)
shapes_valid = shapes[shapes['shape_id'].isin(onibus['shape_id'])]
shapes_valid
```

```python
# Juntando latitude e longitude e convertendo em pontos
shapes_valid = gpd.GeoDataFrame(shapes_valid, geometry=gpd.points_from_xy(shapes_valid.shape_pt_lon, shapes_valid.shape_pt_lat))
shapes_valid
```

```python
from shapely.geometry import Polygon

# Juntando os pontos em polígonos de acordo com o shape_id
shapes_polygon = shapes_valid.groupby('shape_id')['geometry'].apply(lambda x: Polygon(x.tolist())).reset_index()
shapes_polygon
```

```python
import matplotlib.pyplot as plt

# Exemplo de polígono
plt.plot(*shapes_polygon.iloc[673, 1].exterior.xy)
```

```python
# Salvando (fora da pasta, colocar na pasta manualmente)
arquivo = 'shapes.shp'
shapes_polygon.to_file(arquivo)
```
