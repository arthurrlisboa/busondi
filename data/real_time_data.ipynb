```python
def get_real_time_data():
    while(True):
        try:
            real_time_rsp = requests.get(real_time_url).content
            print(real_time_rsp)
            real_time_data = pd.read_csv(io.StringIO(real_time_rsp.decode('utf-8')), delimiter=';', lineterminator='\n')
            return real_time_data
        except:
            sleep(1)
            print('HTTP Error 502: Bad Gateway')
```

```python
print(get_real_time_data())
```

```python
real_time_data = real_time_data.rename(columns=lambda x: x.strip())
```

```python
real_time_data.columns.values
```

```python
bus_data = real_time_data.loc[real_time_data['NL'] == 66]
```

```python
bus_data
```

```python
bus_coords = []
for index, row in bus_data.iterrows():
    lat = float(row['LT'].replace(',', '.'))
    lon = float(row['LG'].replace(',', '.'))
    bus_coords.append((lon, lat))
bus_coords
```
